#Load Libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

#Data Loading

```python
data = pd.read_csv("orbit.csv")
time_steps = data['time_steps'].to_numpy()
y_positions = data['y'].to_numpy()
```

#Model Construction

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()

model.add(Dense(32, input_shape=(1, ), activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(264, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(32, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(16, activation='relu', kernel_regularizer=l2(0.01)))

model.add(Dense(1))

model.summary()
```

#Model Training

```python
# Compile your model
model.compile(optimizer='adamw', loss='mse')
model.fit(time_steps, y_positions, epochs=300)


print('Training started..., this can take a while:')
```

#Predicted orbit vs Scientist's Orbit

```python
def plot_orbit(model_preds):
    axeslim = int(len(model_preds) / 2)
    plt.plot(np.arange(-axeslim, axeslim + 1),np.arange(-axeslim, axeslim + 1) ** 2,
             color="mediumslateblue")
    plt.plot(np.arange(-axeslim, axeslim + 1),model_preds,color="orange")
    plt.axis([-40, 41, -5, 550])
    plt.legend(["Scientist's Orbit", 'Your orbit'],loc="lower left")
    plt.title("Predicted orbit vs Scientist's Orbit")

# Predict the eighty minute orbit
eighty_min_orbit = model.predict(np.arange(-100, 101))

# Plot the eighty minute orbit
plot_orbit(eighty_min_orbit)
```

#Test Evaluation

```python
from sklearn.metrics import mean_absolute_error, r2_score

model_preds = model.predict(np.arange(-100, 101))

actual_orbit = np.arange(-100, 101) ** 2

mse = np.mean((actual_orbit - model_preds.flatten()) ** 2)
mae = mean_absolute_error(actual_orbit, model_preds.flatten())
r_squared = r2_score(actual_orbit, model_preds.flatten())

# Print the metrics
print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'R-squared (R²): {r_squared}')
```
